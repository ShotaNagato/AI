```python
import numpy as np
import matplotlib.pyplot as plt
from classification_clustering_data import * # 外部ファイルからデータ読み込み

# クラスタ数
k = 4
max_roop = 100
color_k = ['#984ea3','#ff7f00','#a65628','#e41a1c']

# ユークリッド距離を計算する関数
def calc_euclidean_distance(point1, point2):
    return np.sqrt((point2[0]-point1[0])**2 + (point2[1]-point1[1])**2)

# クラスタ毎の重心を計算する関数
def calc_center(array_point, array_k):
    # 空の2次元配列を定義
    array_re = np.empty((0,2))
    # クラスタの数繰り返し
    for i in range(k):
        # 重心のXY
        xy_center = array_point[array_k == i].mean(axis=0) # 行方向、すなわち列ごとの平均
        array_re = np.append(array_re, np.array([xy_center]),axis=0) # axis=0で行追加(axis=1なら列追加)
    # 確定した2次元配列を返す
    return array_re

# 各点から最短距離の重心を調べる
def min_distance(array_point, array_center):
    # 空の配列を定義
    array_re = np.empty(0, int)
    # 全ての点に関して繰り返し
    for point in array_point:
        # 空の配列を定義 距離の配列
        array_distance = np.empty(0, int)
        # ある点から全ての重心に向けて距離を計算していく
        for center in array_center:
            array_distance = np.append(array_distance, calc_euclidean_distance(point, center))
        # 距離の配列の中の最小値のところの番号
        array_re = np.append(array_re, array_distance.argmin())
    return array_re

# k-meansでクラスタリング
def kmeans():
    # 種を固定
    np.random.seed(0)
    # 暫定版クラスタの配列
    array_k_preliminary = np.random.randint(0, k, len(STUDENTS)) # 初期状態はランダム
    print("【初期状態のクラスタ】\n"+str(array_k_preliminary))
    show_fig(STUDENTS, array_k_preliminary, color_k)
    # 繰り返しクラスタを更新
    for i in range(max_roop):
        print(str(i+1)+"回目の繰り返しを始めます")
        # 各クラスタごとに重心を計算する
        array_center_preliminary = calc_center(STUDENTS, array_k_preliminary)
        print("【重心】\n"+str(array_center_preliminary))
        # 各点から最短距離となる重心を調べ、各点が属するクラスタを更新
        array_k_update = min_distance(STUDENTS, array_center_preliminary)
        print("【計算し直したクラスタ】\n"+str(array_k_update))
        # 前と同じなら繰り返し終了
        if((array_k_update == array_k_preliminary).all()==True):
            break
        array_k_preliminary = array_k_update
        # 表示
        show_fig(STUDENTS, array_k_preliminary, color_k)
        # show_fig(STUDENTS, array_center_preliminary, color_k)
    # 繰り返し終了、最終結果を返す
    return array_k_preliminary

# 結果を図示
def show_fig(array_point, array_k, array_color):
    plt.figure(figsize=(12, 12))
    plt.title("Students, n=200")
    plt.xlabel("Online")
    plt.ylabel("Offline")
    for i in range(k):
        show_data = array_point[array_k == i]
        plt.scatter(show_data[:,0],show_data[:,1],c=array_color[i],s=60)
        show_center = calc_center(array_point, array_k)
        plt.scatter(show_center[:,0],show_center[:,1],c=array_color[i],s=180)
    plt.show()

###### メイン処理 ######
result_clustering = kmeans()
print("【クラスタリング結果】\n"+str(result_clustering))

show_fig(STUDENTS, result_clustering, color_k)
```
